```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

# For categorical variables

# 1. countplot

```python
sns.countplot(x=titanic['Pclass'])
```

```python
sns.countplot(x=titanic['Survived'])
```

```python
sns.countplot(x=titanic['Embarked'])
```

```python
titanic['Survived'].value_counts().plot(kind='bar')
```

# 2.Pie chart


```python
titanic['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

## For numerical Data

## 1.Histogram

```python
plt.hist(titanic['Fare'])
```

```python
plt.hist(titanic['Age'],bins=5)
```

## 2.Distplot

```python
sns.distplot(x=titanic['Age'])
```

## 3. Boxplot

```python
sns.boxplot(titanic['Age'])
```

```python
print ("mean",titanic['Age'].mean())
print("median",titanic['Age'].mean())
print("mode",titanic['Age'].mean())
print("skew",titanic['Age'].mean())
```
