```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
```

```python
#Trying gradient descent but starting from b=0, we have to reach to b=26.15, trying where are we with b=0
y_pred=X*78.35+0
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='GD')
```

```python
m=78.35
b=0
loss_slope= -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
learnr=0.1
step_size=learnr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='GD')
plt.plot(X,((78.35*X)+20.92),color='pink',label='GD2')
```

```python
#iteration 2
loss_slope= -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=learnr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='GD')
plt.plot(X,((78.35*X)+20.92),color='pink',label='GD2')
plt.plot(X,((78.35*X)+25.11),color='blue',label='GD3')
```

```python
b=-100
epochs=10
m=78.35
learnr=0.1
for i in range(epochs):
    loss_slope= -2*np.sum(y-m*X.ravel()-b)
    step_size=learnr*loss_slope
    b=b-step_size
    y_pred=m*X+b
    plt.plot(X,y_pred)
plt.scatter(X,y)
```
